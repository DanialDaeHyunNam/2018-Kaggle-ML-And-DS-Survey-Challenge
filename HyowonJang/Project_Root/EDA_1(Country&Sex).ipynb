```python
def dfs_sex_per_country(country_name = []):
    dfs = {}
    country_name = country_name
    for i in country_name:
        df = ks.get_df_I_want(i)
        df_male = df[df['Q1']=="Male"]
        df_female = df[df['Q1']=="Female"]
        dfs["{}_male".format(i)] = df_male
        dfs["{}_female".format(i)] = df_female
    return dfs

# TypeError: tuple indices must be integers, not str
# don't make a tuple with stupid commas
```
